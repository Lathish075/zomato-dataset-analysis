# Merging of Data sets

### Top 10 popular cuisines from Zomato 

```python
final_df.columns
```

```python
cuisines_data=final_df[["Country","Cuisines"]].groupby(["Country","Cuisines"]).size().reset_index()
```

```python
cuisine_data
```

```python
cuisines_data = final_df[["Country", "Cuisines"]].groupby(["Country", "Cuisines"]).size().reset_index(name='Count')

# top 10 most popular cuisines globally
top_cuisines = cuisines_data.groupby("Cuisines")["Count"].sum().nlargest(10).index
filtered_data = cuisines_data[cuisines_data["Cuisines"].isin(top_cuisines)]
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Cuisines', hue='Country', data=filtered_data, dodge=False)
plt.title('Top 10 Popular Cuisines Distribution by Country')
plt.xlabel('Count')
plt.ylabel('Cuisine')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
